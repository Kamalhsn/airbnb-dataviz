```python
import pandas as pd

BASE_PATH = "/content/drive/MyDrive/AlmaBetter/Module_4/data"

berlin_listings = pd.read_csv(f"{BASE_PATH}/berlin/listings.csv")
berlin_reviews = pd.read_csv(f"{BASE_PATH}/berlin/reviews.csv")
berlin_neighbourhoods = pd.read_csv(f"{BASE_PATH}/berlin/neighbourhoods.csv")

bangkok_listings = pd.read_csv(f"{BASE_PATH}/bangkok/listings.csv")
bangkok_reviews = pd.read_csv(f"{BASE_PATH}/bangkok/reviews.csv")
bangkok_neighbourhoods = pd.read_csv(f"{BASE_PATH}/bangkok/neighbourhoods.csv")

```

## Schema Inspection



```python
# Schema inspection: shapes and column names

datasets = {
    "berlin_listings": berlin_listings,
    "berlin_reviews": berlin_reviews,
    "berlin_neighbourhoods": berlin_neighbourhoods,
    "bangkok_listings": bangkok_listings,
    "bangkok_reviews": bangkok_reviews,
    "bangkok_neighbourhoods": bangkok_neighbourhoods,
}

for name, df in datasets.items():
    print(f"\n{name}")
    print("Shape:", df.shape)
    print("Columns:", list(df.columns))
```

## Missing Values Overview



```python
# Missing values overview (count only)

for name, df in datasets.items():
    print(f"\n{name}")
    print(df.isna().sum())
```

## Initial Observations

* Both Berlin and Bangkok datasets have the same structure and similar columns, which allows direct comparison between cities.
* Important fields related to pricing, location, and demand are available for analysis. Some columns contain missing values, especially in price, review-related, and license information, which will be handled carefully in the data cleaning stage.
* At this point, no changes have been made to the original data, as the focus is only on understanding data quality and structure.
